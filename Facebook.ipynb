a. Create Data Subsets

```python
subset_1 = df[['like', 'share']]
print(subset_1)
```

```python
subset_2 = df[['comment', 'Type']]
print(subset_2)
```

b. Merge Data

```python
merged_data = pd.merge(subset_1, subset_2, left_on='like', right_on='comment')
```

```python
print(merged_data)
```

c. Sort Data

```python
merged_data.sort_values(by='like', ascending= False)
```

d. Transposing Data

```python
merged_data.transpose()
```

```python
merged_data.T
```
